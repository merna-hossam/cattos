(2)Visualize a 3D data with gray scale, colored and filled contour plot,
X value = 50 samples from 0 to 5, Y value = 90 samples from 0 to 5,
Z value represented by cos(X*Y*100) using meshgrid function to
build 2D grids from 1D arrays and save your figures in .png images

```python
plt.plot(iris.sepal_length, 'go')
plt.plot(iris.sepal_width,'bx')
```

```python
plt.plot(iris.petal_length, 'co')
plt.plot(iris.petal_width,'rx')
```

```python

x=np.linspace(0,5,50,endpoint='True')
y=np.linspace(0,5,90,endpoint='True')
fig=plt.figure(num='Figure One' )
plt.plot(x, 'co')
plt.plot(y,'rx')

```

```python

X,Y = np.meshgrid(x,y)
Z = np.cos(X*Y*100)
fig = plt.figure(figsize=(6,6))
a = fig.add_subplot(111, projection='3d',facecolor='gray')
a.plot_surface(X, Y, Z)
plt.title('fig1')
plt.show()
fig.savefig('fig1.png')
```
